**Imbalanced Dataset**  
A dataset with unequal class distribution

```python
# Importing the dependencies
import pandas as pd
import numpy as np
```

```python
# Loading the dataset to the pandas dataframe
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
# First 5 rows
credit_card_data.head()
```

```python
# Last 5 rows
credit_card_data.tail()
```

```python
# Distribution of 2 data classes
credit_card_data['Class'].value_counts()
```

**This is a Highly Imbalanced Dataset**  
        0.0 --> Legit Transactions  
        1.0 --> Fraud Transactions

```python
# Seperating the legit and fradulent transaction
legit = credit_card_data[credit_card_data.Class == 0.0]
fraud = credit_card_data[credit_card_data.Class == 1.0]
```

```python
print(legit.shape, fraud.shape)
```

**Undersampling :**  
- deletes elements from the majority class to match the number of minority class
<br>
fraud -->156

```python
# Selecting 156 random elements from the legit transaction
legit_sample = legit.sample(n=156)
print(legit_sample.shape)
```

**Concatenate the legit_sample and fraudulent data to form a new dataset**

```python
new_dataset = pd.concat((legit_sample,fraud), axis=0)
new_dataset.head()
```
