**Vector Operations:**
1. Vector Addition
2. Vector Subtraction
3. Vector Multiplication by a Scaler
4. Angle between two Vectors

```python
# Importing the dependencies
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a Vector

```python
plt.quiver(0,0,9,10)
plt.show()
# Shows incorrect graph
```

```python
#Shows correct graph after giving other variables
plt.quiver(0,0,7,5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiple vectors

```python
plt.quiver(0,0,7,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-6,5, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Addition of two vectors :**

*Let's add these two vectors and visualize it*

[ -2, 4 ] + [ 3, 6 ] = [ 1, 10 ]


```python
# Addition of two vectors
vector_1 = np.asarray([0,0,-2,4])
vector_2 = np.asarray([0,0,3,6])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,-2,4, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,6, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,1,10, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-11,11)
plt.ylim(-11,11)
plt.show()
```

**Subtraction of two vectors :**

*Let's subtract these two vectors and visualize it*

[ -2, 4 ] - [ 3, 6 ] = [ -5, -2 ]


```python
# Addition of two vectors
vector_1 = np.asarray([0,0,-2,4])
vector_2 = np.asarray([0,0,3,6])
sum = vector_1 - vector_2
print(sum)
```

```python
plt.quiver(0,0,-2,4, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,6, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,-5,-2, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-11,11)
plt.ylim(-11,11)
plt.show()
```

**Multiplying a Vector with a scaler**

*Let's multiply a vector with a scaler*

[ 3, 2 ] * 2 = [ 6, 4 ]

```python
# Multiplication of vector by a scaler
vector_1 = np.asarray([ 0, 0, 3, 2 ])
vector_2 = 2 * vector_1
print(vector_2)
```

```python
plt.quiver(0,0,3,2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,6,4, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-11,11)
plt.ylim(-11,11)
plt.show()
```

```python
# Multiplication of vector by a decimal
vector_1 = np.asarray([ 0, 0, 3, 2 ])
vector_2 = 0.2 * vector_1
print(vector_2)
```

```python
plt.quiver(0,0,3,2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,0.6,0.4, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```
