```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
Df = pd.read_csv('heart.csv')
Df.head()
```

```python
X = Df.drop(columns='target',axis=1)
Y = Df['target']
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
X_predict = model.predict(X_train)
X_prediction = model.predict(X_test)
```

```python
def precision_recall_f1(Y_true,Y_predict):
    precision = precision_score(Y_true,Y_predict)
    recall = recall_score(Y_true,Y_predict)
    f1 = f1_score(Y_true,Y_predict)
    print('Precision :',precision)
    print('................................................')
    print('Recall :',recall)
    print('................................................')
    print('F1 :',f1)
    print('................................................')
```

```python
# Getting scores for training data
precision_recall_f1(Y_train,X_predict)
```

```python
# Getting scores for testing data
precision_recall_f1(Y_test,X_prediction)
```
