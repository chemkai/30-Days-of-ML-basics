**Statistics**  
* **Types of Data :**
  * Categorical data
    * Nominal data : *Classification of Categorical variables that do not provide any quantitative value. e.g. color of a ball.*
    * Ordinal data : *type of data in which the values follow an actual order. e.g. Hierarchy in education system such as graduate, seconday & elementary etc.*
  * Numerical data
    * Discrete data : *data that have a certain values such as number of employees in the office.*
    * Continuous data : *can have any numerical data between two realistic range. e.g. Weight of the students in a class.*
* **Types of Statistics :**
  * Descriptive Statistics : *used to describe the basic features of the dataset. i.e. it provides simple summary of the dataset.*
  * Inferential Statistics : *takes data from the sample and makes inferences and predictions about the larger population from which the sample was drawn.*
* **Types of Statistical studies :**
  * Sample Study
  * Observational Studies
  * Experimental Studies
* **Types of Sampling Techniques**
  * Simple Random Samples :
    * *Randomly picked deom a larger population.*
    * *All the individual datapoints has an equal probability to be selected as sample data.*
    * *E.g. names of 25 employees being chosen out of a hat from a company of 250 employees.*
  * Systematic Samples :    
    *  *Sample is picked from the population at regular intervals.*
    * *Carried out only if the population is homogeneous and the datapoints are uniformly distributed.*
    * *E.g. Selecting every 10th member from the population.*
  * Stratified Random Sampling :    
    * *Population is subdivided into smaller groups called 'Strata'.*
    * *Samples are obtained randomly from all the Stratas.*
    * *E.g. dividing the population into low, middle, and high-income groups and then selecting a sample from each group.*
  * Clustering Sampling :    
    * *Carried out on population that has inherit groups.*
    * *This population is subdivided into clusters and then random clusters are taken as sample.*
    * *E.g. Smartphone sales in randomly selected states.*

**Central Tendencies in Data Pre-Processing**
* Useful in handling the missing values in a dataset.
* **Mean :**  
  * mean.png
  * *Average of the datapoints in the dataset.*
  * *Missing values can be replaced with the mean value if the data is uniformly distributed.*
* **Median :**
  * *Middle number found by arranging the datapoints in the increasing/decreasing order.*
  * *Missing values can be replaced with median value if the data is skewed.*
* **Mode :**
  * *The value that has the highest frequency in the datapoints.*
  * *Useful for replacement if the datapoints is skewed, especially if the data type is categorical.*

*Standard Deviation :
  * *Square root of the variance.*
